# Log Transformations

```python
import scipy.stats as stat
import pylab 
```

```python
data = df.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(df[feature])
        plt.subplot(1,3,1)
        plt.scatter(data[feature],data['phishing'])
        plt.xlabel(feature)
        
        plt.title(f"{feature} vs phishing")
        plt.subplot(1,3,2)
        stat.probplot(df[feature],dist='norm',plot=pylab)
        
        plt.subplot(1,3,3)
        sns.distplot(data[feature])
        plt.show()
        
```

# Reciprocal Transformation

```python
data = df.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = 1/df[feature]
        plt.subplot(1,3,1)
        plt.scatter(data[feature],data['phishing'])
        plt.xlabel(feature)
        
        plt.title(f"{feature} vs phishing")
        plt.subplot(1,3,2)
        stat.probplot(data[feature],dist='norm',plot=pylab)
        
        plt.subplot(1,3,3)
        sns.distplot(data[feature])
        plt.show()
```

# Square root transformations

```python
data = df.copy()
for feature in continuous_features:
   
    data[feature] = np.sqrt(df[feature])
    plt.subplot(1,3,1)
    plt.scatter(data[feature],data['phishing'])
    plt.xlabel(feature)
        
    plt.title(f"{feature} vs phishing")
    plt.subplot(1,2,2)
    stat.probplot(data[feature],dist='norm',plot=pylab)
    plt.show()
```

# Exponential

```python
data = df.copy()
for feature in continuous_features:
   
    data[feature] = df[feature]**(1/1.2)
    plt.subplot(1,3,1)
    plt.scatter(data[feature],data['phishing'])
    plt.xlabel(feature)
        
    plt.title(f"{feature} vs phishing")
    plt.subplot(1,3,2)
    stat.probplot(data[feature],dist='norm',plot=pylab)
    
    plt.subplot(1,3,3)
    sns.distplot(data[feature])
    
    plt.show()
```

# BoxCOx Transformation

```python
data = df.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        try:
            data[feature], parameters =  stat.boxcox(df[feature])
        except:
            continue
        plt.subplot(1,3,1)
        plt.scatter(data[feature],data['phishing'])
        plt.xlabel(feature)
        
        plt.title(f"{feature} vs phishing")
        plt.subplot(1,3,2)
        stat.probplot(data[feature],dist='norm',plot=pylab)
        
        plt.subplot(1,3,3)
        sns.distplot(data[feature])
        
        plt.show()
```

#### Observation: Clearly box cox transformation seems to be working and making the distributions normal
